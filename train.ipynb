```python
from ultralytics import YOLO
model = YOLO("yolo12s.pt")
results = model.train(
    data='/content/combined/data.yaml',
    epochs=100,
    imgsz=640,
    hsv_h=0.01,
    hsv_s=0.3,
    hsv_v=0.2,
    shear=5,
    perspective=0.0005,
    degrees=180,
    translate=0.1,
    scale=0.5,
    flipud=0.5,
    fliplr=0.5,
    mosaic=1.0,
    mixup=0.15,
    device=0
)
```

```python
results = model(
    source="/content/1.png",
    save=True,
    save_txt=False,
    device=0
)
```
